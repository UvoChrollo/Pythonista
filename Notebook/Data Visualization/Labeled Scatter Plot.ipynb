```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]
y = [item*2 for item in x]

plt.figure(figsize=(8,8))
plt.scatter(x,y,marker='s',s=50)

for x, y in zip(x, y):
    plt.annotate(
        '(%s, %s)' %(x, y), 
        xy=(x, y), 
        xytext=(0, -5),
        textcoords='offset points', 
        ha='center', 
        va='top')
    
plt.grid('y')

plt.show()
```
