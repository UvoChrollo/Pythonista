#### Summary Statistics

```python
homelessness['family_members'].mean()
```

```python
homelessness['family_members'].agg(['mean','median','std'])
```

```python
homelessness[['family_members']].agg(['mean','median','std'])
```

```python
homelessness[['family_members']].agg(['min','max'])
```

```python
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

homelessness[['family_members']].agg(iqr)
```

```python
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

homelessness[['family_members','state_pop']].agg(iqr)
```

```python
import numpy as np

homelessness[['family_members','state_pop']].agg([np.mean,np.median])
```
